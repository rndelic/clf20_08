Используя скаченные данные и `Pandas` (или `PySpark`) посчитайте:
- явку (%) по всем ТИК, результат отсортировать по убыванию
- выберите произвольного кандидата и найдите тот избиратльный участок, на котором он получил наибольший результат  (%, учитывать участки на которых проголосовало больше 100 человек)
- найдите ТИК, где разница между УИК с наибольшей явкой и наименьшей максимальна
- посчитайте дисперсию по явке для каждого ТИК (учитывать УИК)
- для каждого кандидата посчитайте таблицу: результат (%, округленный до целого) - количество УИК, на которых кандидат получил данный результат
- визуализируйте отношение - явка на УИК $\rightarrow$ результат кандидата. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('poster')
```

```python
data = pd.read_csv('voting_data_clean.csv')
data.drop('Unnamed: 0', axis=1, inplace=True)
```

## 1. Явка по ТИК

```python
ax = (data.groupby('tik').sum().num_docs_raw / data.groupby('tik').sum().num_voters_whole).sort_values()[::-1].plot.bar(figsize=(20,10), title='Явка по ТИК')
ax.set_xlabel('ТИК')
ax.set_ylabel('Явка, %')
```

## 2. Максимальный результат Амосова

```python
amosov_results = data[data.num_docs_valid > 100].groupby('tik').sum().Amosov / data[data.num_docs_valid > 100].groupby('tik').sum().num_docs_valid
amosov_results[np.argmax(amosov_results)], amosov_results.index[np.argmax(amosov_results)]
```

## 3. ТИК, где разница между УИК с наибольшей и наименьшей явкой максимальна

```python
data['turnout'] = data.num_docs_given / data.num_voters_whole
maxmin_differences = data.groupby('tik').max().turnout - data.groupby('tik').min().turnout
maxmin_differences[np.argmax(maxmin_differences)], maxmin_differences.index[np.argmax(maxmin_differences)]
```

## 4. Дисперсия явки по каждому ТИК

```python
ax = data.groupby('tik').var().turnout.sort_values()[::-1].plot.bar(figsize=(20,10), title='Дисперсия явки по ТИК')
ax.set_xlabel('ТИК')
ax.set_ylabel('Дисперсия')
```

## 5. Таблица Результат $\rightarrow$ Количество УИК, на которых достигнут результат

```python
data['Amosov_results'] = (data.Amosov / data.num_docs_valid * 100).map(int)
data['Beglov_results'] = (data.Beglov / data.num_docs_valid * 100).map(int)
data['Tikhonova_results'] = (data.Tikhonova / data.num_docs_valid * 100).map(int)

table_results = []
table_results.append(['Amosov', data[data.Amosov_results == 16].shape[0]])
table_results.append(['Beglov', data[data.Beglov_results == 65].shape[0]])
table_results.append(['Tikhonova', data[data.Tikhonova_results == 17].shape[0]])
table_results = pd.DataFrame(table_results)
table_results.columns = ['Candidate', 'num_of_uiks_with_final_result']
```

```python
table_results
```

## 6. Визуализация отношения Явка на УИК $\rightarrow$ Результат кандидата

```python
fig, ax = plt.subplots(3, figsize=(20, 30))
ax[0].scatter(data.turnout, data.Amosov_results)
ax[0].set_xlabel('Явка, %')
ax[0].set_ylabel('Количество голосов')
ax[0].set_title('Амосов')

ax[1].scatter(data.turnout, data.Beglov_results)
ax[1].set_xlabel('Явка, %')
ax[1].set_ylabel('Количество голосов')
ax[1].set_title('Беглов')

ax[2].scatter(data.turnout, data.Tikhonova_results)
ax[2].set_xlabel('Явка, %')
ax[2].set_ylabel('Количество голосов')
ax[2].set_title('Тихонова')

fig.tight_layout()
```
